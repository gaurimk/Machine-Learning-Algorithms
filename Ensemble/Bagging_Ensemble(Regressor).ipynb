```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
```

```python
X=df.iloc[:,0:13]
y=df.iloc[:,13]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,test_size=0.20,random_state=123)
```

```python
print('Train/Test Sets Sizes:',X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
knn=KNeighborsRegressor()
```

```python
lr.fit(X_train,y_train)
dt.fit(X_train,y_train)
knn.fit(X_train,y_train)
```

```python
y_pred1=lr.predict(X_test)
y_pred2=dt.predict(X_test)
y_pred3=knn.predict(X_test)
```

```python
print("R2_score for lr:",r2_score(y_test,y_pred1))
print("R2_score for dt:",r2_score(y_test,y_pred2))
print("R2_score fot knn:",r2_score(y_test,y_pred3))
```

```python
from sklearn.ensemble import BaggingRegressor
```

```python
bag_regressor = BaggingRegressor(random_state=1)
bag_regressor.fit(X_train,y_train)
```

```python
y_preds=bag_regressor.predict(X_test)

print('Training coefficient of r2:%.3f'%bag_regressor.score(X_train,y_train))
print('Test coefficient of r2:%.3f'%bag_regressor.score(X_test,y_test))
```
